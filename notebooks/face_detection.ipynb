```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
image = cv2.imread('/home/sidravic/downloaded_images/v1/f8ce22f3-c905-4400-a668-c84c0abc125f/p435930-av-01-Lhero.jpg')
image2 = cv2.imread('/home/sidravic/downloaded_images/v1/ef2608be-9ac4-4da6-990e-7cb93c44b1f8/beautyblender-re-dew-set-and-refresh-spray-d-2018112007321878~646245_alt5.jpg')

```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=10,
        minSize=(30, 30)
) 

faces2 = faceCascade.detectMultiScale(
        gray2,
        scaleFactor=1.3,
        minNeighbors=10,
        minSize=(30, 30)
) 


print("Found {0} Faces!".format(len(faces)))
print("Found {0} Faces!".format(len(faces2)))

faces2
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(image2, (x, y), (x+w, y+h), (0, 255, 0), 2)
status = cv2.imwrite('faces_detected2.jpg', image2)
status
```
